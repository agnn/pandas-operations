```python
import pandas as pd
```

```python
'Read the agent performace data'

agent_performance = pd.read_csv("F:\ineuron FSDS\pandas practice\Tasks\AgentPerformance (1).xlsx - Agent Performance Report.csv",header = None)
```

```python
agent_performance.columns = agent_performance.iloc[1]
```

```python
agent_performance.drop([0,1], inplace= True)
```

```python
agent_performance
```

```python
agent_performance.columns = agent_performance.columns.str.replace(" ","_") #column header transformation
```

```python
agent_performance
```

```python
#column data type conversions
```

```python
agent_performance['Total_Chats'] = agent_performance.Total_Chats.astype(int)
```

```python
agent_performance[['SL_No','Total_Feedback']] = agent_performance[['SL_No','Total_Feedback']].apply(pd.to_numeric)
```

```python
agent_performance[['Date']]=agent_performance[['Date']].apply(pd.to_datetime)
```

```python
agent_performance['Average_Rating'] = pd.to_numeric(agent_performance['Average_Rating'])
```

```python
agent_performance[['Average_Response_Time','Average_Resolution_Time']]=agent_performance[['Average_Response_Time','Average_Resolution_Time']].apply(pd.to_timedelta)
```

```python
agent_performance # data have been cleaned now
```

```python
agent_performance.dtypes  #converted data types
```

```python
agent_performance.describe()
```

```python
agent_performance['Week'] = agent_performance['Date'].dt.isocalendar().week  #added a week column
```

```python
agent_performance
```

```python
'Now we add a new column Is_working based on assumption that if Total_chats in a day is 0, agent is not wokring'
```

```python
l=[]
for i in (agent_performance['Total_Chats'].values): # appended list by agent 
    if i==0:
        l.append(0)
    else:
        l.append(1)
```

```python
agent_performance['Total_Chats'].values
```

```python
agent_performance['Is_working']=l # assigning the list to the new column
```

```python
agent_performance
```

```python
# Q2 > Total Working days of employees is as below 
'Assumptions taken is if Total_chats is 0, agent is not Working that day'
```

```python
total_working_days =agent_performance.groupby('Agent_Name')['Is_working'].sum()
```

```python
total_working_days
```

```python
working_week_days = agent_performance.groupby(['Agent_Name','Week'])['Is_working'].sum() # Days working per week by agents
```

```python
working_week_days
```

```python
# Q1 - AVERAGE RATING per agent per week based on actual amount of days worked.(This is actual rating, unbiased)
```

```python
agent_performance.groupby(['Agent_Name','Week'])['Average_Rating'].sum()/working_week_days
```
